```python
import os
from PyPDF2 import PdfReader
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt

def extract_text_from_pdf(file_path):
    text = ""
    with open(file_path, "rb") as f:
        pdf_reader = PdfReader(f)
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text

def word_tokenization(text):
    tokens = word_tokenize(text)
    return tokens

def calculate_word_count(tokens):
    return len(tokens)

if __name__ == "__main__":
    # Example usage
    # Define document paths
    documents = {
        "Artificial Intelligence": "/content/Artificial Intelligence (1).pdf",
        "Machine Learning": "/content/Machine Learning.pdf",
        "ADBMS": "/content/ADBMS .pdf",
        "DBMS": "/content/DBMS.pdf",
        "Statistics Probability": "/content/Statistics Probability.pdf",
        "Mathematics": "/content/Mathematics.pdf",
        "Data Communication": "/content/Data communication.pdf",
        "Network": "/content/Network.pdf",
        "Soft Computing": "/content/Soft Computing (1).pdf",
        "Artificial Neural Networks": "/content/Artificial Neural Networks.pdf",
        "JAVA": "/content/JAVA.pdf",
        "Advanced JAVA": "/content/Advanced Java.pdf",
        "Data Structures Using C": "/content/Data structures Using C.pdf",
        "Data Structures Using Python": "/content/Data structures Using Python.pdf",
        "Big Data Analytics": "/content/Big Data Analytics.pdf",
        "Data Mining": "/content/Data Mining.pdf",
        "Embedded System": "/content/Embedded System.pdf",
        "Microprocessor": "/content/Microprocessor.pdf",
        "Ethical Hacking": "/content/Ethical Hacking.pdf",
        "Security": "/content/Security.pdf"
    }

    # Initialize lists to store document names and word counts
    document_names = []
    word_counts = []

    # Iterate over documents
    for doc_name, doc_path in documents.items():
        # Extract text from PDF files
        text = extract_text_from_pdf(doc_path)

        # Perform word tokenization
        tokens = word_tokenization(text)

        # Calculate word count
        count = calculate_word_count(tokens)

        # Append document name and word count to lists
        document_names.append(doc_name)
        word_counts.append(count)

    # Plot the line graph
    plt.figure(figsize=(12, 8))
    plt.plot(document_names, word_counts, marker='o', color='skyblue', linestyle='-')
    plt.xlabel('Document')
    plt.ylabel('Word Count')
    plt.title('Word Count for Each Document')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

 **Similarity between images**

```python
import os
from PyPDF2 import PdfReader
from pdf2image import convert_from_path
import imagehash

def extract_images_from_pdf(file_path):
    images = []
    with open(file_path, "rb") as f:
        pdf_reader = PdfReader(f)
        for page_num, page in enumerate(pdf_reader.pages):
            # Convert page to images
            page_images = convert_from_path(file_path, first_page=page_num+1, last_page=page_num+1)
            images.extend(page_images)
    return images

def compare_images(image_list1, image_list2):
    similar_images = []
    for img1 in image_list1:
        for img2 in image_list2:
            # Calculate the perceptual hash of images
            hash1 = imagehash.average_hash(img1)
            hash2 = imagehash.average_hash(img2)
            # Compare the perceptual hashes
            similarity = 1 - (hash1 - hash2) / len(hash1.hash) ** 2
            # If similarity is above a threshold, consider them similar
            if similarity > 0.9:  # You can adjust the threshold as needed
                similar_images.append((img1, img2, similarity))
    return similar_images

if __name__ == "__main__":
    # Example usage
    Machine_Learning1_doc_path = "/content/Machine Learning1.pdf"
    Machine_Learning2_doc_path = "/content/Machine Learning2.pdf"

    # Extract images from the PDF documents
    images_ml1 = extract_images_from_pdf(Machine_Learning1_doc_path)
    images_ml2 = extract_images_from_pdf(Machine_Learning2_doc_path)

    # Compare the images and display similar ones
    similar_images = compare_images(images_ml1, images_ml2)

    # Display the similar images and their similarity percentage
    for img_pair in similar_images:
        img1, img2, similarity = img_pair
        print(f"Similarity between images: {similarity:.2%}")
```
